```python
size = [179, 172,28,22,17,12,11,9,7,6,6,6,4,1 ]
country = ['KuwaIT', 'lebanon', 'Egypt', 'SaudiArabia', 'USA', 'Jordan', 'venzuela', 'Iran',
 'Tunis', 'Morocco', 'Syria', 'Palestine', 'Iraq', 'Lybia']
sizes = size
print(sizes)
labels = list(country)
print(labels)
cols = ['red','green','blue','yellow','gold','yellowgreen','lightcoral','cyan','lightskyblue','blue',
        'orange','purple','pink','indigo']
plt.pie(sizes, labels = labels,  colors = cols, startangle = 90, shadow = True, explode = None, 
        frame =True,autopct = "%0.1f%%")
plt.rcParams["figure.figsize"] = (15,18)
plt.title('Country distribution')
plt.axis('equal')
plt.legend()
plt.show()
```
